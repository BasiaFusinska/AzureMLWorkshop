Set up workspace environment.

```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
from azureml import Workspace
ws = Workspace()
workdspace_id = ws.workspace_id
authorization_token = ws.authorization_token
```

Get the dataset.

```python
iris = datasets.load_iris()
```

Fit the model.

```python
from sklearn.linear_model import LogisticRegression

X = iris.data
y = iris.target

lm = LogisticRegression()
lm.fit(X, y)
```

Predict some values.

```python
print(lm.predict(X[10]))
print(lm.predict(X[60]))
print(lm.predict(X[100]))
```

Publish a web service.

```python
from azureml import services

@services.publish(workdspace_id, authorization_token)
@services.types(sl=float, sw=float, pl=float, pw=float)
@services.returns(float)

def predict_species(sl, sw, pl, pw):
    return lm.predict([sl, sw, pl, pw])
```

Use the service.

```python
predict_species.service(X[60][0], X[60][1], X[60][2], X[60][3])
```
